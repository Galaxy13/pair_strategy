```python
from datetime import datetime
import json
from pyravendb.store import document_store
from pyravendb.raven_operations.timeseries_operations import *
from pyravendb.data.query import IndexQuery
from pyravendb.commands.raven_commands import *
from pyravendb.data.document_conventions import DocumentConventions
from pyravendb.data.timeseries import TimeSeriesRange

from_date_var = datetime(year=2023, month=3, day=1, hour=12, minute=0, second=0)
to_date_var = datetime(year=2023, month=3, day=2, hour=12, minute=0, second=0)

time_series_range = TimeSeriesRange("Candles",from_date_var, to_date_var)

with document_store.DocumentStore(urls=["http://89.223.124.208:8080"], database="MarketData") as store:
    store.initialize()
    requests_executor = store.get_request_executor()
    time_series = store.operations.send(GetTimeSeriesOperation("83225_5", ranges=time_series_range))
    query_command = QueryCommand(
        index_query=IndexQuery(query="From 'FinamSecurityDatas' where Id()='83225_5'",
                                wait_for_non_stale_results=True), 
                                conventions=DocumentConventions())
    response = requests_executor.execute(query_command)

    
    
    
    #with store.open_session() as session:
    #    query_result = list(session.query().where_equals("Key", "83225_5"))

time_series_s = json.dumps(time_series)
time_series_data = json.loads(time_series_s)
print(time_series_data)
print(time_series_data['Entries'])
print(time_series_data['Entries'][0])
print(time_series_data['Entries'][1]['Values'])
print(time_series_data['Entries'][1]['Values'][0])
print("-------------")

response_s = json.dumps(response)
response_data = json.loads(response_s)
print(response_data['Results'])
print(response_data['Results'][0])
print(response_data['Results'][0]['ClassName'])
print(response_data['Results'][0]['Code'])
print(response_data['Results'][0]['Em'])
```
